```python
#Importing the libraries

import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url="https://www.amazon.com/s?k=sarees&ref=nb_sb_noss_1"
driver.get(url)
```

```python
saree=[]

for j in range(0,6):

    s=driver.find_elements_by_xpath("//div[@class='a-section aok-relative s-image-square-aspect']")
    for i in s:
        sa=i.find_element_by_tag_name("img")
        sarees = saree.append(sa.get_attribute('src'))
    button=driver.find_element_by_xpath("//*[contains(text(), 'Next')]")
    driver.execute_script("arguments[0].click();", button)
    time.sleep(5)
saree
```

```python
len(saree)
```

```python
import requests
for img in saree:
 file1 = img.split('/')[-1]
 print("Downloading file:%s"%file1)
 r = requests.get(img, stream=True)
 with open(file1, 'wb') as f:
    for chunk in r:
       f.write(chunk)
```
